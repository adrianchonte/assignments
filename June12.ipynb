```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
customers = pd.read_csv('customers-1.csv')
```

```python
products = pd.read_csv('products.csv')
```

```python
sales = pd.read_csv('sales.csv')
```

```python
customers['registration_date'] = pd.to_datetime(customers ['registration_date'])
```

```python
products['launch_date'] = pd.to_datetime(products ['launch_date'])
```

```python
sales['transaction_date'] = pd.to_datetime(sales ['transaction_date'])
```

```python
customers['reg_day'] = customers['registration_date'].dt.day_name()
customers['reg_year'] = customers['registration_date'].dt.year
customers['reg_month'] = customers['registration_date'].dt.month_name()
customers
```

```python
sales['trans_day'] = sales['transaction_date'].dt.day_name()
sales['trans_year'] = sales['transaction_date'].dt.year
sales['trans_month'] = sales['transaction_date'].dt.month_name()
sales
```

```python
products['launch_day'] = products['launch_date'].dt.day_name()
products['launch_year'] = products['launch_date'].dt.year
products['launch_month'] = products['launch_date'].dt.month_name()
products
```

```python
merged = pd.merge(sales, products, on='product_id', how='left')
print(merged.head())
```

```python
# Create the 2x2 plot grid
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Revenue by Category
sns.barplot(data=revenue_by_category.sort_values(by='revenue'), x='revenue', y='category', ax=ax1)
ax1.set_title('Revenue by Category')

# Plot 2: Top 6 States
sns.barplot(data=top_states, x='state', y='customer_count', ax=ax2)
ax2.set_title('Top 6 States by Customer Count')

# Plot 3: Avg Transaction by Day
avg_daily = merged.groupby('transaction_date')['revenue'].mean().reset_index()
ax3.plot(avg_daily['transaction_date'], avg_daily['revenue'])
ax3.set_title('Average Transaction Amount by Day')

# Plot 4: Histogram
sns.histplot(merged['revenue'], bins=30, ax=ax4)
ax4.axvline(merged['revenue'].mean(), color='red', linestyle='--', label='Mean')
ax4.set_title('Distribution of Transaction Amounts')
ax4.legend()

fig.suptitle('Sales Analysis Dashboard', fontsize=16)  # Add main title with font size

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to leave space for the main title
plt.tight_layout()
plt.show()

# Display everything

```

```python
# Group by customer_id, aggregate sum and count of total_amount
customer_spending = merged.groupby('customer_id')['total_amount'].agg(['sum', 'count']).reset_index()

# Rename columns to clearer names
customer_spending.rename(columns={'sum': 'total_spent', 'count': 'transaction_count'}, inplace=True)

# Check the result
print(customer_spending.head())
```

```python
# Merge spending summary with customer details on 'customer_id'
customer_data = pd.merge(customer_spending, customers, on='customer_id', how='left')
```

```python
# Define bins and labels for age groups
bins = [0, 29, 45, 60, 100]
labels = ['Under 30', '30-45', '46-60', '60+']

# Create a new column 'age_group'
customer_data['age_group'] = pd.cut(customer_data['age'], bins=bins, labels=labels, right=True)
```

```python
print(customer_data[['customer_id', 'age', 'age_group', 'total_spent', 'transaction_count']].head())
```

```python
plt.figure(figsize=(10, 6))

# Scatter plot of income vs total spending
sns.scatterplot(data=customer_data, x='income', y='total_spent', alpha=0.6)

# Add a regression line to show trend
sns.regplot(data=customer_data, x='income', y='total_spent', scatter=False, color='red')

plt.title('Customer Income vs Total Spending')
plt.xlabel('Income')
plt.ylabel('Total Spending')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter plot
sns.scatterplot(data=customer_data, x='income', y='total_spent', alpha=0.6, ax=ax)

# Regression line
sns.regplot(data=customer_data, x='income', y='total_spent', scatter=False, ax=ax, color='red')

# Reference lines (example thresholds, adjust as needed)
avg_income = customer_data['income'].mean()
avg_spent = customer_data['total_spent'].mean()

ax.axvline(avg_income, color='blue', linestyle='--', label='Avg Income')
ax.axhline(avg_spent, color='green', linestyle='--', label='Avg Spending')

# Annotations
ax.text(avg_income + 1000, ax.get_ylim()[1]*0.9, 'Average Income', color='blue')
ax.text(ax.get_xlim()[0]*1.1, avg_spent + 500, 'Average Spending', color='green')

# Optional: annotate a specific interesting point (e.g., highest spender)
top_customer = customer_data.loc[customer_data['total_spent'].idxmax()]
ax.annotate(
    f"Top Spender\nIncome: ${top_customer['income']}\nSpent: ${top_customer['total_spent']}",
    xy=(top_customer['income'], top_customer['total_spent']),
    xytext=(top_customer['income']+5000, top_customer['total_spent']+5000),
    arrowprops=dict(facecolor='black', arrowstyle='->'),
    fontsize=9,
    backgroundcolor='yellow'
)

ax.set_title('Customer Income vs Total Spending with Annotations')
ax.set_xlabel('Income')
ax.set_ylabel('Total Spending')

ax.legend()
plt.show()
```
